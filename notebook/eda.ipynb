```python
#Task B
df.info() 
print("-----------")
df.isna().sum()
```

##### GPA is the column that has the most missing values because GPA is more sensitive information. 

```python
#Task C
import pandas as pd
#count, mean, median, variance, SD, min, max. 
summary = [
    ['Count', df['age'].count(), df['gpa'].count()],
    ['Mean', df['age'].mean(), df['gpa'].mean()],
    ['Median', df['age'].median(), df['gpa'].median()],
    ['Variance', df['age'].var(), df['gpa'].var()],
    ['SD', df['age'].std(), df['gpa'].std()],
    ['Min', df['age'].min(), df['gpa'].min()],
    ['Max', df['age'].max(), df['gpa'].max()]
]

columns = ['statistics', 'age', 'gpa']


summary = pd.DataFrame(summary)
print(summary)
```

```python
#Task D
df['age'].hist(bins=range(14, 19), align='left', rwidth=0.8)
import matplotlib.pyplot as plt; plt.figure()
import seaborn as sns; sns.histplot(df["gpa"], kde=True)
```

```python
#Task E: Explore relationships
import matplotlib.pyplot as plt
plt.scatter(df["age"],df["gpa"])
plt.xlabel("Age")
plt.ylabel("GPA")
df["age"].corr(df["gpa"])
```
